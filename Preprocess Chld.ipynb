```python
import pandas as pd

df = pd.read_csv('dataset/data_balita.csv')
df = df.iloc[:, 1:]
df.value_counts('Status Gizi')
```

```python
status_gizi = {
    'normal' : 1,
    'stunted' : 2,
    'tinggi' : 3,
    'severely stunted': 4
}

jenis_kel = {
    'laki-laki': 0,
    'perempuan' : 1
}
df['Status Gizi'] = df['Status Gizi'].map(status_gizi)
df['Jenis Kelamin'] = df['Jenis Kelamin'].map(jenis_kel)
df
```

```python
df['Protein_g'] = (((df['Tinggi Badan (cm)'] / 5) *  df['Status Gizi'] / 5) + (df['Umur (bulan)']*2))
df
```

```python
df['Energy_kcal'] = (((df['Tinggi Badan (cm)'] / 5) *  df['Status Gizi'] / 5)) + ((df['Umur (bulan)']+1)*65)
df
```

```python
umur = df['Umur (bulan)'] <= 11

df.loc[umur].sort_values('Umur (bulan)')
```

```python
df['Calcium_mg'] = 1000+ 20**(df['Status Gizi']-1)
df
```

```python
food_df_new = pd.read_csv('~/capstone2/dataset/preprocessed_dataset.csv')
food_df = food_df_new.drop(['Jumlah Porsi', 'Takaran Porsi'], axis=1)
food_df = food_df.loc[:, ['Protein (g)', 'Kalori (kkal)']]
food_df
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import train_test_split
X, y = train_test_split(food_df , shuffle=False, train_size=0.8)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
prep_data=scaler.fit_transform(X.to_numpy())
```

```python
from sklearn.neighbors import NearestNeighbors

model = NearestNeighbors(metric='cosine',)
model.fit(prep_data)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer
transformer = FunctionTransformer(model.kneighbors,kw_args={'return_distance':False})
pipeline=Pipeline([('std_scaler',scaler),('NN',transformer)])
```

```python
params={'n_neighbors':5,'return_distance':False}
# pipeline.get_params()
pipeline.set_params(NN__kw_args=params)
```

```python
pred_idx = 5
input = y.iloc[pred_idx]
prediction = pipeline.transform([input])[0]
food_df_new.iloc[prediction]
# preds = food_df_new.iloc[pred.index].sort_values('Protein (g)', ascending=False)
```

```python
y.iloc[pred_idx]
```

```python
import pickle


filename = 'recommender_model.h5'
knnPickle = open(filename, 'wb') 
      
# source, destination 
pickle.dump(pipeline, knnPickle)  

# close the file
knnPickle.close()
```
